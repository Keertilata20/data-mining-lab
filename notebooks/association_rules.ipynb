```python
import pandas as pd

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
transactions = [
    ['milk', 'bread', 'butter'],
    ['milk', 'bread'],
    ['milk', 'bread', 'eggs'],
    ['bread', 'butter'],
    ['milk', 'eggs'],
    ['bread', 'eggs'],
    ['milk', 'bread', 'butter', 'eggs'],
    ['butter', 'eggs'],

    ['beer', 'chips'],
    ['beer', 'chips', 'nuts'],
    ['beer', 'nuts'],
    ['chips', 'nuts'],
    ['beer', 'chips'],

    ['rice', 'dal', 'oil'],
    ['rice', 'dal'],
    ['oil', 'spices'],
    ['rice', 'oil'],
    ['dal', 'spices'],
    ['rice', 'dal', 'oil', 'spices'],

    ['milk', 'cereal'],
    ['cereal', 'bread'],
    ['milk', 'cereal', 'bread'],
    ['cereal', 'butter'],

    ['eggs', 'bread', 'butter'],
    ['milk', 'bread', 'chips'],
    ['beer', 'bread'],
    ['milk', 'chips'],
    ['bread', 'chips'],
]
```

```python
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_data, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
rules.sort_values(by='lift', ascending=False).head(10)
```

## 📊 Observations

- The association rules reveal strong relationships between commonly purchased items.
- The rule milk → bread has higher confidence compared to bread → milk.
- Lift values greater than 1 indicate positive associations between items.
- The dataset successfully captures realistic shopping patterns.

## 🧠 Insight

- Association rule mining identifies directional relationships between items.
- Milk is a stronger predictor of bread purchases than vice versa.
- Lift values confirm that these relationships are not random but meaningful.
- Such insights can be used in real-world applications like product placement and recommendation systems.
